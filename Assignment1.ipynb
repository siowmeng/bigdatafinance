```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics, cross_validation
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score

from scipy.stats import pearsonr
```

```python
loans = pd.read_excel('LCloanbook.xls')
loans_Y = loans["loan_status_CO_or_Default_or_Late"]
loans_unscaled = loans.drop("loan_status_CO_or_Default_or_Late", axis = 1)
loans_X = MinMaxScaler().fit_transform(loans_unscaled)
loans_X = pd.DataFrame(loans_X)
```

```python
loans_unscaled.head()
```

```python
loans.corr().sort(columns = "loan_status_CO_or_Default_or_Late")
```

```python
loans_full_logistic = cross_val_score(LogisticRegression(), loans_X, loans_Y, scoring = "accuracy", cv = 10)
loans_full_logistic.mean()
```

```python
loans_full_tree = cross_val_score(DecisionTreeClassifier(), loans_X, loans_Y, scoring = "accuracy", cv = 10)
loans_full_tree.mean()
```

```python
loans_full_knn = cross_val_score(KNeighborsClassifier(n_neighbors = 1), loans_X, loans_Y, scoring = "accuracy", cv = 10)
loans_full_knn.mean()
```
